```python
import numpy as np
import tensorflow as tf
import tensorflow_probability as tfp
tfd = tfp.distributions
```

Using the example from the tutorial: https://www.tensorflow.org/probability/api_docs/python/tfp/distributions/StudentT

```python
df_input = tf.constant(2.0)
loc_input = tf.constant(2.0)
scale_input = tf.constant(11.0)
```

```python
dist = tfd.StudentT(df=df_input, loc=loc_input, scale=scale_input)
```

*sample(n)* function is part of Distribution class that calls the sample_n specified by inherited Distribution class.

```python
sample = dist.sample(sample_shape=1000).numpy()
```
