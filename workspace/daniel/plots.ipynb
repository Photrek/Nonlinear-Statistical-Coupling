**CoupledProbability**

```python
def CoupledNormalDistribution(mean, sigma, kappa, alpha):
    """
    Short description
    
    Inputs
    ----------
    x : Input variable in which the coupled logarithm is applied to.
    mean : 
    sigma : 
    kappa : Coupling parameter which modifies the coupled logarithm function.
    dim : The dimension of x, or rank if x is a tensor. Not needed?
    """

    assert sigma >= 0, "std must be greater than or equal to 0."
    assert alpha in [1, 2], "alpha must be set to either 1 or 2."

    coupledNormalDistributionResult = []
    if kappa >= 0:
        input = np.arange(mean-20, mean+20, (20+mean - -20+mean)/16385)
    else:
        x1 = mean - ((-1*sigma**2) / kappa)**0.5
        x2 = mean + ((-1*sigma**2) / kappa)**0.5
        input = np.arange(mean - ((-1*sigma**2) / kappa)**0.5, mean + ((-1*sigma**2) / kappa)**0.5, (x2-x1)/16385)
 
    normCGvalue = 1/float(normCG(sigma, kappa))
    for i in input:
        coupledNormalDistributionResult.append(normCGvalue * (CoupledExponential((i - mean)**2/sigma**2, kappa)) ** -0.5)
  
    return coupledNormalDistributionResult
```

```python
result = CoupledNormalDistribution(0, 1, 0, 1)
x = np.arange(-20, 20, (20 - -20)/10000)
plt.plot(x, result)
plt.show()
```
